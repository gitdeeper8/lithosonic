```python
import numpy as np
import matplotlib.pyplot as plt

# Campi Flegrei parameters 2023-2024
dates = ['2023-01', '2023-06', '2024-01', '2024-06', '2024-11']
b_c = [0.57, 0.63, 0.71, 0.78, 0.81]

plt.figure(figsize=(10, 6))
plt.plot(dates, b_c, 'o-', linewidth=2, markersize=8)
plt.axhline(y=0.75, color='orange', linestyle='--', label='Elevated Threshold')
plt.ylabel('B_c (Biot Coupling)')
plt.title('Campi Flegrei - Biot Coupling Evolution 2023-2024')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

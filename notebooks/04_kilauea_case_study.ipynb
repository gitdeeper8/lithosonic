# Kīlauea 2018 Eruption Case Study
## LITHO-SONIC Tutorial 4

```python
import numpy as np
import matplotlib.pyplot as plt
from litho_physics.lsi import compute_lsi

# Simulated LSI leading up to 2018 eruption
days = np.arange(-90, 10)
lsi = 0.4 + 0.004 * days + 0.5 * np.exp((days + 30) / 30)
lsi = np.clip(lsi, 0, 1)

plt.figure(figsize=(10, 6))
plt.plot(days, lsi, 'b-', linewidth=2)
plt.axhline(y=0.60, color='orange', linestyle='--', label='Elevated Alert (0.60)')
plt.axhline(y=0.80, color='red', linestyle='--', label='Critical (0.80)')
plt.axvline(x=-14, color='red', alpha=0.3, label='LSI Critical - Day -14')
plt.xlabel('Days before eruption')
plt.ylabel('LSI')
plt.title('Kīlauea 2018 Eruption - LSI Evolution')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Simulated ROC curve
np.random.seed(42)
n_samples = 1000

# Generate synthetic scores
background = np.random.normal(0.4, 0.15, n_samples)
events = np.random.normal(0.75, 0.15, n_samples)

scores = np.concatenate([background, events])
labels = np.concatenate([np.zeros(n_samples), np.ones(n_samples)])

fpr, tpr, _ = roc_curve(labels, scores)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 8))
plt.plot(fpr, tpr, 'b-', linewidth=2, label=f'ROC (AUC = {roc_auc:.3f})')
plt.plot([0, 1], [0, 1], 'k--', alpha=0.5)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('LITHO-SONIC Detection Performance')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

print(f'Area under ROC curve: {roc_auc:.3f}')
print(f'Expected AUC from paper: 0.97 ± 0.02')
```

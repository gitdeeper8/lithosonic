# The Geysers Induced Seismicity Case Study

```python
import numpy as np
import matplotlib.pyplot as plt

# Injection and seismicity correlation
injection_rate = [800, 950, 1100, 1250, 1400, 1300, 1150, 1000]
seismicity = [2, 5, 12, 25, 42, 35, 18, 8]

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

ax1.plot(injection_rate, 'b-', linewidth=2)
ax1.set_ylabel('Injection Rate (L/min)')
ax1.set_title('The Geysers - Injection and Induced Seismicity')
ax1.grid(True, alpha=0.3)

ax2.plot(seismicity, 'r-', linewidth=2)
ax2.set_xlabel('Time (months)')
ax2.set_ylabel('M > 2.5 Events per month')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

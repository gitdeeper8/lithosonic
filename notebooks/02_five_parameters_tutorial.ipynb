## 1. Biot Coupling Coefficient (B_c)

```python
props = RockProperties(
    porosity=0.20,
    permeability=1e-13,
    bulk_modulus_frame=5e9,
    bulk_modulus_grain=40e9,
    bulk_modulus_fluid=2.2e9,
    density_grain=2700,
    density_fluid=1000,
    viscosity_fluid=1e-3
)

solver = BiotSolver()
b_c = solver.compute_biot_coupling(props)
print(f'B_c = {b_c:.3f}')
```
